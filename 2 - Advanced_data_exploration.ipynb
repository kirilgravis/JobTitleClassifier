# Building some functions for data processing notebook:

* nltk library   - - -  building the function

```python
from nltk.stem import WordNetLemmatizer

def nltk_lemmatization_func(row):
    nltk_lemmatizer = WordNetLemmatizer()
    return [nltk_lemmatizer.lemmatize(w) for w in row]
```

```python
df['nltk_lemmatization'] = df.Desc.apply(nltk_lemmatization_func)
df.head(2)
```

* textblob library  - - -  building the function

```python
from textblob import Word

def textblob_lemmatization_func(row):
    return [Word(w).lemmatize() for w in row]
```

```python
df['textblob_lemmatization'] = df.Desc.apply(textblob_lemmatization_func)
df.head(2)
```

* spacy library  - - -  building the function

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def spacy_lemmatization_func(row):
    words_spacy = " ".join([w for w in row])
    doc_words = nlp(words_spacy)
    tokens_words = []
    for token in doc_words:
        tokens_words.append(token)
    return [token.lemma_ for token in tokens_words]
```

```python
df['spacy_lemmatization'] = df.Desc.apply(spacy_lemmatization_func)
df.head(2)
```

* stemming ---- building the function

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def stemming_func(row):
    stemming = PorterStemmer()
    return [stemming.stem(w) for w in row]
```

```python
df['stemming'] = df.Desc.apply(stemming_func)
df.head(2)
```

* stemming after spacy lemmatization

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def stemming_func(row):
    stemming = PorterStemmer()
    return [stemming.stem(w) for w in row]
```

```python
df['stemming_after_spacy_lemmatization'] = df.spacy_lemmatization.apply(stemming_func)
df.head(2)
```

```python
df.to_csv('processed_round1_plus_three_types_of_lemmatization_plus_stemming_plus_stemming_on_spacy_lemmatization.csv', index=False)
```

#### List of tags

```python
from textblob import TextBlob

def POS_tagging(text):    
    text = " ".join([w for w in text])
    return TextBlob(text).tags

def noun_amount(text):
    noun_list = ['NN','NNS','NNP','NNPS']
    text = " ".join([w for w in text])
    count = 0
    for w in TextBlob(text).tags:
        if w[1] in noun_list:
            count += 1
    return count

def noun_only(text):
    noun_list = ['NN','NNS','NNP','NNPS']
    text = " ".join([w for w in text])
    l1 = []
    for w in TextBlob(text).tags:
        if w[1] in noun_list:
            l1.append(w[0])
    return l1

round2 = lambda x: POS_tagging(x)
round3 = lambda x: noun_only(x)
```
